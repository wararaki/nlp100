```python
import pickle

from sklearn.metrics import accuracy_score
import torch
from tqdm import tqdm
```

```python
def load(filename: str) -> tuple:
    with open(filename, 'rb') as f:
        (X, y) = pickle.load(f)
    X = torch.from_numpy(X.astype(float)).float()
    y = torch.from_numpy(y.astype(float)).long()
    return (X, y)
```

```python
filename = 'data/train.pkl'
(X_train, y_train) = load(filename)
```

```python
filename = 'data/valid.pkl'
(X_valid, y_valid) = load(filename)
```

```python
class NNModel(torch.nn.Module):
    def __init__(self, n_input: int, n_output: int):
        super(NNModel, self).__init__()
        self.linear_1 = torch.nn.Linear(n_input, n_output)
        self.softmax = torch.nn.Softmax(dim=1)
    
    def forward(self, x: torch.Tensor) -> float:
        x = self.linear_1(x)
        return self.softmax(x)
```

```python
model = NNModel(300, 4)
criterion = torch.nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)
```

```python
model.train()
train_losses = []
valid_losses = []
train_accuracies = []
valid_accuracies = []

epochs = 5
for epoch in tqdm(range(1, epochs+1)):
    model.train()
    for x, y in zip(X_train, y_train.reshape(-1, 1)):
        optimizer.zero_grad()
        y_pred = model(x.reshape(1, -1))
        loss = criterion(y_pred, y)
        loss.backward()
        optimizer.step()

    model.eval()
    with torch.no_grad():
        y_pred = model(X_train)
        loss = criterion(y_pred, y_train)
        accuracy = (y_pred.argmax(dim=1) == y_train).sum().item()/y_train.shape[0]
        train_losses.append(loss.item())
        train_accuracies.append(accuracy)
    
    with torch.no_grad():
        y_pred = model(X_valid)
        loss = criterion(y_pred, y_valid)
        accuracy = (y_pred.argmax(dim=1) == y_valid).sum().item()/y_valid.shape[0]
        valid_losses.append(loss.item())
        valid_accuracies.append(accuracy)
    
    torch.save({
        'epoch': epoch,
        'model_state_dict': model.state_dict(),
        'optimizer_state_dict': optimizer.state_dict(),
        'loss': loss
    }, f'models/model_{epoch}.pkl')
```

```python
import matplotlib.pyplot as plt
```

```python
fix, ax = plt.subplots()
ax.plot(list(range(1, 6)), train_accuracies, label='train_accuracy')
ax.plot(list(range(1, 6)), valid_accuracies, label='valid_accuracy')
plt.show()
```

```python
fix, ax = plt.subplots()
ax.plot(list(range(1, 6)), train_losses, label='train_loss')
ax.plot(list(range(1, 6)), valid_losses, label='valid_loss')
plt.show()
```
